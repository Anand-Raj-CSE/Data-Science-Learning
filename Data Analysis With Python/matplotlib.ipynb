```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# creating a line plot
plt.plot(x,y)
```

```python
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('This is a simple line plot.')
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red',linestyle='--',marker = 'o',markersize=9)
plt.grid(True)
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]
plt.figure(figsize=(9,5))

plt.subplot(2,2,1) # 1 row and 2 column, it is used when we have multiple plots
plt.plot(x,y1,color='green')
plt.title('Simplefigure')

plt.subplot(2,2,2) 
plt.plot(x,y1,color='red')
plt.title('Simplefigure2')

plt.subplot(2,2,3) # 1 row and 2 column
plt.plot(x,y2,color='blue')
plt.title('third box')
```

### Bar plot

```python
categories = ['A','B','C','D','E']
values = [5,7,3,8,6]

# creating a bar plot
plt.bar(categories,values,color='maroon')
```

### Histogram -  it divides data into bins and count the number of data points in each bin

```python
data = [1,1,1,2,2,3,3,5,6,6,9,8,7,8,3,21,5,11,2,2,21,9,1,10]
plt.hist(data,color='orange',edgecolor='black')
```

```python
#sctter plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.scatter(x,y,marker='x')
```

```python
#pipe chart
labels = ['a','b','c','d','e']
sizes = [30,20,40,10,50]
colors = ['gold','yellowgreen','blue','lightskyblue','navy']
explode = (0.1,0,0,0,0) # move ou tthe first slice
plt.pie(sizes,labels=labels,colors=colors,shadow=True,explode=explode)
```

### Real time data usage

```python
import pandas as pd
```

```python
sales_data = pd.read_csv('sales_data.csv')
sales_data.head()
```

```python
#Plot total sales by product
total_sales_grp_by = sales_data.groupby('Product Category')['Total Revenue'].sum()
print(total_sales_grp_by)
```

```python
total_sales_grp_by.plot(kind='bar',color='teal')
```

```python
sales_trend = sales_data.groupby('Date')['Total Revenue'].sum()
sales_trend
```

```python
sales_trend = sales_data.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend # Now we are getting two columns
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
