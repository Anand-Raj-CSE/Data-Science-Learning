```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# creating a line plot
plt.plot(x,y)
```

```python
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('This is a simple line plot.')
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red',linestyle='--',marker = 'o',markersize=9)
plt.grid(True)
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]
plt.figure(figsize=(9,5))

plt.subplot(1,2,1) # 1 row and 2 column, it is used when we have multiple plots
plt.plot(x,y1,color='green')
plt.title('Simplefigure')

plt.subplot(2,2,2) # 1 row and 2 column
plt.plot(x,y1,color='green')
plt.title('Simplefigure')
```
