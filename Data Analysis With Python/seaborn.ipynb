```python
import seaborn as sns
```

### Basic plotting with seaborn

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot for tips")
plt.show()
```

```python
#Lineplot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Total bill vs size")
plt.show()
```

```python
# Categorical plots
#Bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bills acccording to days')
plt.show()
```

```python
# Box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True) # kde gives the curve of distribution , kde is kernel density estimator.
plt.show()
```

```python
# KDE plot - kernel density estimator
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# pair plot = we want to campare features then we use this , it only takes features that have numerica value
sns.pairplot(tips)
```
