# Sets - they donot allow duplicates elements and are unordered.


```python
# Set creation
my_set = {1,2,3,4}
print(my_set)
```

```python
my_set = set([1,2,3,4,5,4,5,6,7,7,5,67,4,3,67,3])
print(my_set)
```

```python
# Whever we try to remove any element from thr set whivh is not present .
my_set.remove(10)
```

```python
#For this we use
my_set.discard(10)
```

```python
print(my_set)
my_set.add(10)
removed_element = my_set.pop()  # pop uses first in first out
print(removed_element)
print(my_set)
```

```python
# Set membership test
print(10 in my_set)
```

# mathematical operations in sets

```python
# set unuion
set_one = {1,2,3,4,5,6,7,8,9,123,34,45,78,56}
set_two = {12,123,45,56,78,90,87}
union_set = set_one.union(set_two)
print(union_set)

intersection_set = union_set.intersection(set_two)
print(intersection_set)
```

### intersection_update , it would update set1 with the common elements , it will delete all the previous elements, there would be update , in all , difference update , differentiation update and all , all the update methods would be update the original variable with the result.

```python
# intersection_update , it would update set1 with the common elements , it will delete all the previous elements
set_one.intersection_update(set_two)
print(set_one)
```

```python
set_one = {1,2,3,4,5,6,7,8,9,123,34,45,78,56}
set_two = {12,123,45,56,78,90,87}
```

```python
print(set_one.difference(set_two))
```

# symmetric difference , all the duplicate elements would be removed and the unique elements will be there

```python
set_one = {1,2,3,4,5,6,7,8,9,123,34,45,78,56}
set_two = {12,123,45,56,78,90,87}
```

```python
print(set_one.symmetric_difference(set_two))
```

```python
set_one = {1,2,3,4,5,6,7,8,9,123,34,45,78,56}
set_two = {12,123,45,56,78,90,87}
print(set_one.issubset(set_two))
print(set_one.issuperset(set_two))
```
